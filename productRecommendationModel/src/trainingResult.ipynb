```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def show_result(path, save_path):
    plt.style.use('default')  # Reset do domyślnego stylu (z białym tłem)
    plt.rcParams['figure.facecolor'] = 'white'  # Tło figury
    plt.rcParams['axes.facecolor'] = 'white'   # Tło osi
    history = pd.read_csv(path)
    plt.figure(figsize=(23, 8))
    plt.subplot(1, 2, 1)
    
    plt.plot(history['loss'], label='Strata treningowa')
    plt.plot(history['val_loss'], label='Strata walidacyjna')
    plt.title('Zmiana straty w czasie trenowania')
    plt.xlabel('Epoka')
    plt.ylabel('Strata')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history['accuracy'], label='Dokładność treningowa')
    plt.plot(history['val_accuracy'], label='Dokładność walidacyjna')
    plt.title('Zmiana dokładności w czasie trenowania')
    plt.xlabel('Epoka')
    plt.ylabel('Dokładność')
    plt.legend()
    
    if save_path is not None:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')  # dpi ustawia jakość, bbox_inches zmniejsza marginesy
    plt.show()

def one_plot_loss_function(path, save_path):
    history = pd.read_csv(path)
    plt.figure()
    
    plt.plot(history['loss'], label='Strata treningowa')
    plt.plot(history['val_loss'], label='Strata walidacyjna')
    plt.title('Zmiana straty w czasie trenowania')
    plt.xlabel('Epoka')
    plt.ylabel('Strata')
    plt.legend()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    
def one_plot_accuracy(path, save_path):
    history = pd.read_csv(path)
    plt.figure()
    
    plt.plot(history['accuracy'], label='Dokładność treningowa')
    plt.plot(history['val_accuracy'], label='Dokładność walidacyjna')
    plt.title('Zmiana dokładności w czasie trenowania')
    plt.xlabel('Epoka')
    plt.ylabel('Dokładność')
    plt.legend()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
```

```python

one_plot_loss_function('history_training/predict_category_1.csv', 'plots/to_documents/loss_1.png')
one_plot_accuracy("history_training/predict_category_1.csv", "plots/to_documents/accuracy_1.png")

```
