```python
def get_list_browse_before_purchasing(one_visitorid_who_bought: pd.DataFrame) -> list : 
    number_items_viewed = 0
    transaction_time = None
    time_threshold = 86400000  # one day 
    result = []
    for i, (index, row) in enumerate(one_visitorid_who_bought.iterrows()):
        if transaction_time:
            if transaction_time - row['timestamp'] < time_threshold and row['event'] != 'transaction':
                number_items_viewed += 1
            else:
                result.append((transaction_time,row['visitorid'], number_items_viewed))
                
            if i == len(one_visitorid_who_bought) - 1:
                result.append((transaction_time, row['visitorid'], number_items_viewed))

        if row['event'] == 'transaction':
            number_items_viewed = 0
            transaction_time = row['timestamp']
    return result


previously_viewed = []
for visitor in events['visitorid'].unique()[200000:300000]:
    one_visitorid_who_bought = events[events['visitorid'] == visitor].sort_values(by='visitorid', ascending=True)

    all_purchase = one_visitorid_who_bought[one_visitorid_who_bought["event"] == "transaction"]

    one_visitorid_who_bought = one_visitorid_who_bought.sort_values(by='timestamp', ascending=False)

    previously_viewed += get_list_browse_before_purchasing(one_visitorid_who_bought)

print(previously_viewed)

one_visitorid_who_bought['timeinterval2'] = pd.to_timedelta(one_visitorid_who_bought['timestamp'], unit='ms')
print(one_visitorid_who_bought)     

```
